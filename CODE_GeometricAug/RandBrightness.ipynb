**RandBrightness**

```python
import os
import cv2
import numpy as np
from google.colab import drive
```

```python
def random_brightness_transform(image):
    """
    Aplica un ajuste aleatorio de brillo a una imagen.
    """
    # Generar un valor de brillo aleatorio entre -30 y 30
    brightness = np.random.randint(-30, 31)

    # Aplicar el ajuste de brillo
    bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness)

    return bright_image, brightness

def generate_output_filename_brightness(original_filename, brightness):
    """
    Genera el nombre del archivo de salida incluyendo el valor de brillo aplicado.
    """
    # Separar el nombre del archivo de su extensión
    name, ext = os.path.splitext(original_filename)
    # Crear nuevo nombre con el valor de brillo
    new_name = f"{name}_brightness_{brightness}{ext}"
    return new_name

# Montar Google Drive
drive.mount('/content/drive')

# Especificar la ruta de la carpeta de entrada y salida
input_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/DATA_AUGM/DATA'  # Cambia esto por la ruta a tu carpeta
output_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/DATA_AUGM/Brightness_imagesRand'

# Crear carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Procesar imágenes
valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
processed_count = 0

# Listar todos los archivos en la carpeta de entrada
for filename in os.listdir(input_folder):
    if any(filename.lower().endswith(ext) for ext in valid_extensions):
        # Ruta completa de la imagen
        input_path = os.path.join(input_folder, filename)

        # Leer la imagen
        img = cv2.imread(input_path)

        if img is not None:
            # Aplicar transformación
            transformed_img, brightness = random_brightness_transform(img)

            # Generar nuevo nombre de archivo con valor de brillo aplicado
            output_filename = generate_output_filename_brightness(filename, brightness)
            output_path = os.path.join(output_folder, output_filename)

            # Guardar imagen transformada
            cv2.imwrite(output_path, transformed_img)

            # Mostrar información
            print(f"\nProcesando: {filename}")
            print(f"Guardado como: {output_filename}")
            print(f"Dimensiones originales: {img.shape}")
            print(f"Ajuste de brillo: {brightness}")

            processed_count += 1

print(f"\nTotal de imágenes procesadas: {processed_count}")
print(f"Las imágenes procesadas se encuentran en: {output_folder}")
```
