**RandSaturation**

```python
import os
import cv2
import numpy as np
from google.colab import drive
```

```python
def random_saturation_transform(image):
    """
    Aplica un ajuste aleatorio de saturación (Saturation) a una imagen.
    """
    # Convertir la imagen al espacio de color HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Separar los canales H, S y V
    h, s, v = cv2.split(hsv_image)

    # Generar un valor de ajuste aleatorio para la saturación
    saturation_shift = np.random.randint(-30, 31)

    # Aplicar el ajuste de saturación
    s = cv2.add(s, saturation_shift)
    s = np.clip(s, 0, 255)  # Asegurarse de que los valores estén en el rango permitido

    # Recomponer la imagen con los canales ajustados
    adjusted_hsv = cv2.merge((h, s, v))
    adjusted_image = cv2.cvtColor(adjusted_hsv, cv2.COLOR_HSV2BGR)

    return adjusted_image, saturation_shift

def generate_output_filename_saturation(original_filename, saturation_shift):
    """
    Genera el nombre del archivo de salida incluyendo el valor de ajuste de saturación aplicado.
    """
    # Separar el nombre del archivo de su extensión
    name, ext = os.path.splitext(original_filename)
    # Crear nuevo nombre con el valor de ajuste de saturación
    new_name = f"{name}_saturation_{saturation_shift}{ext}"
    return new_name

# Montar Google Drive
drive.mount('/content/drive')

# Especificar la ruta de la carpeta de entrada y salida
input_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/DATA_AUGM/DATA'  # Cambia esto por la ruta a tu carpeta
output_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/DATA_AUGM/Saturation_imagesRand2'

# Crear carpeta de salida si no existe
os.makedirs(output_folder, exist_ok=True)

# Procesar imágenes
valid_extensions = ['.jpg', '.jpeg', '.png', '.bmp']
processed_count = 0

# Listar todos los archivos en la carpeta de entrada
for filename in os.listdir(input_folder):
    if any(filename.lower().endswith(ext) for ext in valid_extensions):
        # Ruta completa de la imagen
        input_path = os.path.join(input_folder, filename)

        # Leer la imagen
        img = cv2.imread(input_path)

        if img is not None:
            # Aplicar transformación
            transformed_img, saturation_shift = random_saturation_transform(img)

            # Generar nuevo nombre de archivo con valor de ajuste de saturación aplicado
            output_filename = generate_output_filename_saturation(filename, saturation_shift)
            output_path = os.path.join(output_folder, output_filename)

            # Guardar imagen transformada
            cv2.imwrite(output_path, transformed_img)

            # Mostrar información
            print(f"\nProcesando: {filename}")
            print(f"Guardado como: {output_filename}")
            print(f"Dimensiones originales: {img.shape}")
            print(f"Ajuste de saturación: {saturation_shift}")

            processed_count += 1

print(f"\nTotal de imágenes procesadas: {processed_count}")
print(f"Las imágenes procesadas se encuentran en: {output_folder}")
```
