```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from scipy.fftpack import fft
import os
from google.colab import drive

drive.mount('/content/drive')  # ðŸ”— Montar Google Drive

# ðŸ“Œ ParÃ¡metros del audio
duration = 2  # DuraciÃ³n en segundos
fs = 44100  # Frecuencia de muestreo (Hz)
t = np.linspace(0, duration, int(fs * duration), endpoint=False)  # Vector de tiempo

# ðŸ“Œ Generar seÃ±al con dos frecuencias distintas
f1, f2 = 440, 880  # Frecuencias en Hz (1er seg: 440 Hz, 2do seg: 880 Hz)
signal = np.where(t < 1, np.sin(2 * np.pi * f1 * t), np.sin(2 * np.pi * f2 * t))

# ðŸ”„ Normalizar a 16 bits
signal_int16 = np.int16(signal * 32767)

# ðŸ’¾ Guardar el audio
path = "/content/drive/MyDrive/Data_Augmentation/JUAN/AUDIO_DATA_AUGM/AudioPuro"
os.makedirs(path, exist_ok=True)  # Crear carpeta si no existe
filename = f"{path}/audio_puro_2s.wav"
wav.write(filename, fs, signal_int16)
print(f"âœ… Audio guardado en: {filename}")

# ðŸ“Š Aplicar Transformada de Fourier
yf = fft(signal)  # FFT de la seÃ±al
xf = np.fft.fftfreq(len(yf), 1/fs)  # Escala de frecuencia correcta

# ðŸ” Mostrar resultados
def plot_signal_and_fft():
    plt.figure(figsize=(12, 6))

    # ðŸ”¹ SeÃ±al en el dominio del tiempo
    plt.subplot(2, 1, 1)
    plt.plot(t, signal, label='SeÃ±al en el tiempo', color='b')
    plt.axvline(1, color='r', linestyle='--', label='Cambio de frecuencia')
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.title("SeÃ±al generada")
    plt.legend()

    # ðŸ”¹ Espectro de frecuencia
    plt.subplot(2, 1, 2)
    plt.plot(xf[:len(xf)//2], np.abs(yf[:len(yf)//2]) / len(yf), color='g')  # Magnitud de la FFT
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Magnitud")
    plt.title("Transformada de Fourier (FFT)")
    plt.xlim(0, 2000)  # Ajustar lÃ­mites para ver mejor las frecuencias
    plt.grid()

    plt.tight_layout()
    plt.show()

plot_signal_and_fft()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
from scipy.fftpack import fft
import os
from google.colab import drive

drive.mount('/content/drive')  # ðŸ”— Montar Google Drive

# ðŸ“Œ ParÃ¡metros del audio
duration = 2  # DuraciÃ³n en segundos
fs = 44100  # Frecuencia de muestreo (Hz)
t = np.linspace(0, duration, int(fs * duration), endpoint=False)  # Vector de tiempo

# ðŸ“Œ Generar seÃ±al con oscilaciones caÃ³ticas y calmadas que pueden superar [-1,1]
f1, f2 = 440, 880  # Frecuencias base en Hz
chaotic_oscillation = 50 * np.sin(2 * np.pi * 10 * t)  # Oscilaciones rÃ¡pidas en el primer segundo
calm_oscillation = 15 * np.sin(2 * np.pi * 2 * t)  # Oscilaciones suaves en el segundo segundo

signal = np.where(t < 1, 1.5 * np.sin(2 * np.pi * (f1 + chaotic_oscillation) * t), 1.8 * np.sin(2 * np.pi * (f2 + calm_oscillation) * t))

# ðŸ”„ Normalizar a 16 bits manteniendo la amplitud relativa
max_val = np.max(np.abs(signal))
signal_int16 = np.int16((signal / max_val) * 32767)

# ðŸ’¾ Guardar el audio
path = "/content/drive/MyDrive/Data_Augmentation/JUAN/AUDIO_DATA_AUGM/AudioPuro"
os.makedirs(path, exist_ok=True)  # Crear carpeta si no existe
filename = f"{path}/audio_OOOooo_2s.wav"
wav.write(filename, fs, signal_int16)
print(f"âœ… Audio guardado en: {filename}")

# ðŸ“Š Aplicar Transformada de Fourier
yf = fft(signal)  # FFT de la seÃ±al
xf = np.fft.fftfreq(len(yf), 1/fs)  # Escala de frecuencia correcta

# ðŸ” Mostrar resultados
def plot_signal_and_fft():
    plt.figure(figsize=(12, 6))

    # ðŸ”¹ SeÃ±al en el dominio del tiempo
    plt.subplot(2, 1, 1)
    plt.plot(t, signal, label='SeÃ±al en el tiempo', color='b')
    plt.axvline(1, color='r', linestyle='--', label='Cambio de frecuencia')
    plt.xlabel("Tiempo [s]")
    plt.ylabel("Amplitud")
    plt.title("SeÃ±al generada")
    plt.legend()

    # ðŸ”¹ Espectro de frecuencia
    plt.subplot(2, 1, 2)
    plt.plot(xf[:len(xf)//2], np.abs(yf[:len(yf)//2]) / len(yf), color='g')  # Magnitud de la FFT
    plt.xlabel("Frecuencia [Hz]")
    plt.ylabel("Magnitud")
    plt.title("Transformada de Fourier (FFT)")
    plt.xlim(0, 2000)  # Ajustar lÃ­mites para ver mejor las frecuencias
    plt.grid()

    plt.tight_layout()
    plt.show()

plot_signal_and_fft()
```
