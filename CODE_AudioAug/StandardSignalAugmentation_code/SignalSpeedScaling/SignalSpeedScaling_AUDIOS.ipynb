```python
# Código para generar los audios con modificaciones de velocidad

# Instalar las dependencias
!pip install librosa soundfile

# Importar librerías
import librosa
import numpy as np
import os
import soundfile as sf  # Usar soundfile para manejar archivos multicanal
from google.colab import drive

drive.mount('/content/drive')

# Configuración de rutas
input_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/AUDIO_DATA_AUGM/DATA'  # Ruta a la carpeta de entrada
output_folder = '/content/drive/MyDrive/Data_Augmentation/JUAN/AUDIO_DATA_AUGM/RESULTS/StandardSignalAugmentation/SignalSpeedScaling/AUDIOS' # Carpeta de salida
os.makedirs(output_folder, exist_ok=True)

# Función para modificar la velocidad
def speed_scaling(audio, speed_factor):
    return librosa.effects.time_stretch(audio, rate=speed_factor)

# Procesar los archivos
for file_name in os.listdir(input_folder):
    if file_name.endswith('.wav'):
        print(f"Procesando: {file_name}")
        file_path = os.path.join(input_folder, file_name)

        # Generar un valor aleatorio para el speed_factor
        speed_factor = round(np.random.uniform(0.8, 1.2), 2)  # Rango ajustable

        # Cargar el audio en formato multicanal
        audio, sr = librosa.load(file_path, sr=None, mono=False)  # Conservar múltiples canales

        # Si el audio es multicanal, procesar cada canal por separado
        if len(audio.shape) > 1:  # Audio multicanal
            processed_audio = np.array([speed_scaling(channel, speed_factor=speed_factor) for channel in audio])
            processed_audio = processed_audio.T  # Transponer para mantener el formato multicanal
        else:  # Audio mono
            processed_audio = speed_scaling(audio, speed_factor=speed_factor)

        # Guardar el archivo procesado
        output_path = os.path.join(output_folder, f"speed_sf{speed_factor}_{file_name}")
        sf.write(output_path, processed_audio, sr)  # Guardar usando soundfile

print("Archivos procesados con Signal Speed Scaling conservando múltiples canales.")
```
